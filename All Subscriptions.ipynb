```python
import pandas as pd
data = pd.read_csv("all_subscriptions_mar2.csv")
data = data.fillna('')
```

```python
#For inactive subscriptions, how many times did they check out a box? 

print(len(data))
def getCheckouts(val):
    count = 0
    totalInactive = 0
    for index in range(0, len(data)):
        isActive = data["is_active"][index]
        
        if (not isActive):
            totalInactive = totalInactive + 1
        
        if (not isActive)and(data["total_checkouts"][index] == val):
            count = count + 1
    return totalInactive

# for x in range(0,50):
#     print("{}: {}".format(x,getCheckouts(x)))
    
print(getCheckouts(5))
        
```

```python
#Active Subscriptions

def getCheckoutsActive(val):
    count = 0
    totalActive = 0
    for index in range(0, len(data)):
        isActive = data["is_active"][index]
        
        if isActive:
            totalActive = totalActive + 1
        
        if (isActive)and(data["total_checkouts"][index] == val):
            count = count + 1
    return count

cap = 10
count = 0
for x in range(0,200):
    if(x >= cap):
        count = count + getCheckoutsActive(x)
    print("{}".format(getCheckoutsActive(x)))
    
```
